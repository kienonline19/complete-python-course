## Cho dữ liệu ArticlesApril2017.csv thực hiện các yêu cầu sau:

1. Đọc dữ liệu và tạo đoạn text từ cột headline. Sau đó thực hiện chuẩn hóa đoạn text (loại bỏ các từ 
không quan trọng)
Vd : Các từ không quan trọng ['one', 'br', 'Po', 'th', 'sayi', 'fo', 'Unknown'] 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('ArticlesApril2017.csv')
```

```python
df
```

```python
stopwords = ['one', 'br', 'Po', 'th', 'sayi', 'fo', 'Unknown']
res = df['headline'].apply(lambda sentence: ' '.join(word for word in sentence.split() if word not in stopwords))
res
```

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud
word_cloud = WordCloud(background_color='white').generate(str(res.values))
plt.figure(figsize=(15, 10))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
